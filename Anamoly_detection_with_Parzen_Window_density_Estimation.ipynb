```python
from sklearn.neighbors import KernelDensity
from numpy import where, random, array, quantile
from sklearn.preprocessing import scale
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston
import pandas as pd 
import numpy as np
import os
```

```python
random.seed(124)
def makeData(N):
    x = []
    for i in range(N):
        a = i/1000 + random.uniform(-3, 2)
        r = random.uniform(-5, 10)
        if(r >= 9.8):
            r = r + 10
        elif(r<(-4.8)):
            r = r +(-10)        
        x.append([a + r])   
    return array(x)

n = 500
x= makeData(n)
print(x.shape)
x_ax = range(n)
plt.plot(x_ax, x)
plt.show() 

kernaldens = KernelDensity().fit(x)
scores = kernaldens.score_samples(x)

thresh = quantile(scores, .01)
print(thresh)

index = where(scores <= thresh)
values = x[index]

plt.plot(x_ax, x)
plt.scatter(index,values, color='r')
plt.show()
```

```python
def detect_anamoly(user, i, title, user_File_name):
    
#     userfile=pd.read_csv(user)
    
    arr=user.iloc[:, [i]].to_numpy()

    kernel=KernelDensity().fit(arr)

    score_samples=kernel.score_samples(arr)
    print('score shape:', score_samples.shape)

    threshold = quantile(score_samples, .02)
    index = where(score_samples <= threshold)
    values = arr[index]
#     print('values: ', values[0:9])
    x_ax=range(arr.shape[0])
    index=np.array(index)
    
    print(len(index[0]), values.shape)
    plt.title(f'{title} of user {user_File_name}')
    plt.plot(x_ax, arr)
    plt.scatter(index,values, color='r')
    plt.show()

```
